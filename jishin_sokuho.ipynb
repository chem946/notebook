```python
from BeautifulSoup import BeautifulSoup
import urllib2
# Web からデータを取得する
URL="http://www.jma.go.jp/jp/quake/"
html = urllib2.urlopen(URL).read()
soup = BeautifulSoup(html)
```

```python
#txt
import re
p = re.compile(r"<[^>]*?>")
print p.sub("",soup('table',{'id':'infobox'})[0].prettify())
```
